### Visualização com PCA

## Vetor de tamanho 50 (tokens)

```python
TAMANHO_MINIMO_PALAVRA = 4
QTD_PALAVRAS = 1000

counter = Counter([w for w in list(model50.wv.key_to_index) if len(w) >= TAMANHO_MINIMO_PALAVRA])

pca = PCA(n_components=3)

palavras_mais_freq = [w for w, f in counter.most_common(QTD_PALAVRAS)]

result = pca.fit_transform(model50.wv.vectors_for_all(keys=palavras_mais_freq).vectors)

fig = px.scatter_3d(
    result, x=0, y=1, z=2, 
    text=palavras_mais_freq
)

fig.show();
```
